```python
import nltk
import re
import pandas as pd

msg = pd.read_csv(r"C:\Users\rauna\codes\datasets\SMSSpamCollection", sep ="\t", names = ["label","message"])
```

```python
msg.head()
```

```python
msg["label"].value_counts()
```

```python
from nltk import WordNetLemmatizer
lm = WordNetLemmatizer()
from nltk.corpus import stopwords
```

```python
corpus = []

for i in range(0,len(msg)):
    review = re.sub('[^a-zA-Z]',' ', msg["message"][i])
    review = review.lower()
    review = review.split()
    review = [lm.lemmatize(word) for word in review if not word in stopwords.words("english")]
    review = " ".join(review)
    corpus.append(review)
    
```

```python
corpus
```

## Creating bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
x = cv.fit_transform(corpus).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=30,linewidth=100000,
                    formatter=dict(float=lambda x: "%.3g" % x))
x
```

```python
cv.vocabulary_
```

```python
y = pd.get_dummies(msg["label"],dtype=int)
y = y.iloc[:,0].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect = MultinomialNB().fit(x_train,y_train)
```

```python
spam_detect
```

```python
y_pred = spam_detect.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
acc = accuracy_score
```

```python
acc(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

## TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=2500, ngram_range=(1,2))
```

```python
x = tfidf.fit_transform(corpus)
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
spam_detect_tfidf = MultinomialNB().fit(x_train,y_train)
```

```python
y_pred = spam_detect_tfidf.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

We have done our modelling after transforming the data which causes the data leakage issue. So, we will now try to do the same process but we will do the train_test_split first and then transform them into BOW or TF-IDF

```python
x_train,x_test, y_train,y_test = train_test_split(corpus,y,test_size=0.3)
```

```python
# we will use bow because we were getting more accuracy compare to tf-idf to check if there is any overfitting.
x_train = cv.fit_transform(x_train).toarray()
x_test = cv.transform(x_test).toarray()
```

```python
spam_detect_bow_v2 = MultinomialNB().fit(x_train,y_train)
```

```python
y_pred = spam_detect_bow_v2.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
